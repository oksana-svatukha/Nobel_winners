```python
import pandas as pd
nobel_winners = pd.read_csv('/home/oksana/Jupyter_projects/ Nobel_prize_winners/archive/nobel_laureates_data.csv')
nobel_winners.head()
```

1. Виведи кількість лауреатів за кожною категорією премії

```python
count_nobel_winners = nobel_winners.groupby('category')['fullName'].count().reset_index()
count_nobel_winners
```

2. Вивеи список лауреатів з України

```python
x = nobel_winners[nobel_winners['bornCountry'] == 'Ukraine']
x[['fullName']]
```

3. Виведи список лауреатів, які отримали премію більше одного разу

```python
unique_name = nobel_winners.groupby('fullName').filter(lambda x: len(x) > 1)
df2 = unique_name[['fullName']]
df2.fullName.unique()
```

4. Виведи список лауреатів, які отримали премію в різних категоріях.

```python
count_nobel_winners = nobel_winners.groupby('fullName').filter(lambda x: x['category'].nunique() > 1)
df2 = count_nobel_winners["fullName"]
df2.unique()
```

5. Виведи список лауреатів жінок та впорядку за роком отримання премії.

```python
women_nobel_winners = nobel_winners.loc[nobel_winners['gender'] == 'female'].sort_values(by="year")
women_nobel_winners
```

6. Кільксть жінок які здобули нагороду в 2023 році

```python
women_winners_2023 = women_nobel_winners[women_nobel_winners['year'] == 2023]
women_winners_2023['fullName']
```

7. Виведи які країни отримували більше 50 нагород

```python
country = nobel_winners['bornCountry'].value_counts()
filtered_country = country[country > 50]
filtered_country
```
